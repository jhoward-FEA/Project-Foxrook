```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt, dask.dataframe as dd, pyarrow as pa
import chardet, os, re, xlrd
from openpyxl import load_workbook
```

```python
def get_sheetnames_xlsx(filepath):
    wb = load_workbook(filepath, read_only=True, keep_links=False)
    return wb.sheetnames

sheetnames = get_sheetnames_xlsx('Vision/Vision_BD_SQL.xlsx')
```

```python
dfs = []
for sheet in sheetnames:
    dfs.append(pd.read_excel('Vision/Vision_BD_SQL.xlsx', sheet_name=sheet))
```

```python
for df in dfs:
    print(df.info())
```

```python
df_names = ['AR', 'CL', 'EMProjectAssoc', 'LD', 'LedgerAR', 'Opportunity',
            'OpportunityCustomTabFields', 'PR', 'Projects_BillingProjections', 
            'Projects_ProjectSetup']
for i, df in enumerate(dfs):
    working_filename = str(df_names[i]) + '.pkl'
    df.to_pickle(working_filename)
```
